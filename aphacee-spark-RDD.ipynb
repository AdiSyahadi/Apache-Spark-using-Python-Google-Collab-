## RDD

```python
sc = spark.sparkContext
```

```python
A = [1,2,3,4]
B = ["new york", "tokyo", "sidney", "jakarta", "singapore"]
```

### Merubah ke RDD

```python
rdd_A = sc.parallelize(A)
```

```python
rdd_A = sc.parallelize(B)
```

## RDD Transformation Map

```python
rdd_cities = sc.parallelize(["new york", "tokyo", "sidney", "jakarta", "singapore"])
```

```python
rdd_cities_upper = rdd_cities.map(lambda x: x.upper())
```

### RDD Action : Colect and Count

```python
rdd_cities_upper.collect()
```

```python
rdd_cities_upper.count()
```

## RDD Transformation : Union
Menggabungkan RDD 

```python
rdd_cities2 = sc.parallelize(["san francisco", "delhi"])
```

```python
united_cities = rdd_cities.union(rdd_cities2)
```

```python
united_cities.collect()
```

## RDD ACtion : Take

```python
united_cities.take(3)
```
